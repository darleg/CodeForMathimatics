# Distance Between Two Points

```python
from numpy import linspace
from pylab import figure, subplots, axis, plot, text, title, show

x = linspace(-5,5,100)
fig = figure()
ax = fig.add_subplot()
ax.set_title(' Distance Between Two Points')
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.set_xlim(-5,5)
ax.set_ylim(-5,5)

ax.set_xticks(linspace(-5,5,11))
ax.set_yticks(linspace(-5,5,11))
ax.set_aspect('equal')
ax.grid(True,ls=':')

p1 = [1, 2]
p2 = [3, 4]
x1 = [p1[0], p2[0]]
y1 = [p1[1], p2[1]]
plot(x1, y1, 'bo', linestyle = '--')
text(p1[0]-0.8, p1[1]+0.25, p1)
text(p2[0]-0.8, p2[1]+0.25, p2)



vx2 = [0, 3]
vy2 = [-1, -1]
plot(vx2, vy2, 'b4', linestyle = ':')
text(1.5, -1.4,'x2')
vx1 = [0, 1]
vy1 = [-2, -2]
plot(vx1, vy1, 'r4', linestyle = ':')
text(.4, -2.4,'x1')

text(1.8,-3,'x=x2-x1')

vx4 = [-2, -2]
vy4 = [0, 2]
plot(vx4, vy4, 'r2', linestyle = ':')
text(-1.7, 2,'y1')
vx3 = [-1, -1]
vy3 = [0, 4]
plot(vx3, vy3, 'b2', linestyle = ':')
text(-2.7, 1,'y2')

text(-3,3,'y=y2-y1')

text(1.3,1.1, ' X and Y Distance')

show()
```

```python
from numpy import linspace
from pylab import figure, subplots, axis, plot, text, title, show

x = linspace(-5,5,100)
fig = figure()
ax = fig.add_subplot()
ax.set_title('Distance Formula')
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.set_xlim(-5,5)
ax.set_ylim(-5,5)

ax.set_xticks(linspace(-5,5,11))
ax.set_yticks(linspace(-5,5,11))
ax.set_aspect('equal')
ax.grid(True,ls=':')

p1 = [1, 2]
p2 = [3, 4]
p3 = [3, 2]
x1 = [p1[0], p2[0]]
y1 = [p1[1], p2[1]]
x2 = [p1[0], p3[0]]
y2 = [p1[1], p3[1]]
x3 = [p3[0], p2[0]]
y3 = [p3[1], p2[1]]
plot(x1, y1, 'bo', linestyle = '--')
plot(x2, y2, 'gx', linestyle = ':')
plot(x3, y3, 'g', linestyle = ':')
text(p1[0]-0.8, p1[1]+0.25, p1)
text(p2[0]-0.8, p2[1]+0.25, p2)

text(1.8,1.5,'x=x2-x1')
text(3.1,3,'y=y2-y1')

vx2 = [0, 3]
vy2 = [-1, -1]
plot(vx2, vy2, 'b4', linestyle = ':')
text(1.5, -1.4,'x2')
vx1 = [0, 1]
vy1 = [-2, -2]
plot(vx1, vy1, 'r4', linestyle = ':')
text(.4, -2.4,'x1')

vx4 = [-2, -2]
vy4 = [0, 2]
plot(vx4, vy4, 'r2', linestyle = ':')
text(-1.7, 2,'y1')
vx3 = [-1, -1]
vy3 = [0, 4]
plot(vx3, vy3, 'b2', linestyle = ':')
text(-2.7, 1,'y2')

text(1.3,1.1,'d=sqrt(x**2 + y**2)')

show()
```

Pythagoras theorem for the △ABC:

$AB^2 = AC^2 + BC^2$

$d = \sqrt{(x2-x1)^2 + (y2-y1)^2}$

OR

$d = \sqrt{x^2 + y^2}$
