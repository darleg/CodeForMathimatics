## Coordinate Plane
- The plane space is into two dimensions and is useful to easily locate the points.
- In this case it is also referred to as a cartesian plane.
- Has two axes on the plane which are the horizontal x-axis and the vertical y-axis.
- Axes divide the plane into four quadrants.
- Point of intersection of these axes is the origin (0, 0).
- Any point in the plane is referred to by a point (x, y).
- The x value is the position of the point with reference to the x-axis.
- The y value is the position of the point with reference to the y-axis.

```python
from numpy import linspace
from pylab import figure, subplots, axis, plot, text, show

x = linspace(-5,5,100)

fig = figure()
ax = fig.add_subplot()
ax.set_title('Coordinate Plane')


ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.set_xlim(-5,5)
ax.set_ylim(-5,5)

ax.set_xticks(linspace(-5,5,11))
ax.set_yticks(linspace(-5,5,11))
ax.set_aspect('equal')
ax.grid(True,ls=':')
text(1.3,2.5,'Quardrant I')
text(-3.7,2.5,'Quardrant II')
text(-3.7,-2.5,'Quardrant III')
text(1.3,-2.5,'Quardrant IV')
plot()
show()
```

## Formulas for two Points
- Distance Formula
- Slope Formula
- Mid-Point Formula

### Distance Formula

```python
from math import sqrt

def distance(x1,y1,x2,y2):
    print("distance formula solver")
    return sqrt(((x1-x2)**2) + ((y1-y2)**2))

x1=1
y1=2
x2=3
y2=4

d = distance(x1,y1,x2,y2)
print('Distance = %f' %(d))
```

### Slope Formula

```python
def slopee(x1,y1,x2,y2):
    x = (y2 - y1) / (x2 - x1)
    return x

print(slopee(1,2,3,4))
```

### Mid-Point Formula

```python
def midpoint(x1, y1, x2, y2):
    return ((x1 + x2)/2, (y1 + y2)/2)

print(midpoint(1,2,3,4))
```
