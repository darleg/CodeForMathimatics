<h1 style="text-align: center;"> Linear Equation in One Variable <h1>

```python
from numpy import linspace
from pylab import figure, subplots, axis, plot, title, show
```

<h1 style="text-align: center;"> Generl Linear Equation <h1>
$$y = ax + b$$

```python
fig = figure()
ax = fig.add_subplot()
x = linspace(-5,5,100)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plot(x, 2*x+2, 'r', label='y=2x+2')
plot(x, 2*x-2, 'g', label='y=2x-2')
plot(x, -2*x+3, 'b', label='y=-2x+3')
plot(x, -2*x-3, 'm', label='y=-2x-3')
fig.legend(loc='upper left')
show()
```

<h1 style="text-align: center;"> Slope-Intercept Equation <h1>
$$y=mx + b$$  
<p> m is called the slope <p>
<p> b is defined as the y-intercept <p>
<p> y-intercept the point the line crosses Y-axis <p> 

```python
def plotspec(ax):
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

def Line(m, b):
    y = m * x + b
    return y

def LinePlot(m, b, c, t):
    y = Line(m, b)
    plot(x,y, c)
    title(t)

# min and max x
xl = -5
xr =  5

# Limits of x with 100 spaced number
x = linspace(xl,xr,100)


# setting the axes at the centre
fig = figure()

# first line function which is y = mx + b
ax = fig.add_subplot(2, 2, 1)
plotspec(ax)
m = 2
b = 2
c = 'r'
t = 'y = 2x + 2'
LinePlot(m, b, c, t)

# second line function which is y = mx + b
ax = fig.add_subplot(2, 2, 2)
plotspec(ax)
m = 2
b = -2
c = 'g'
t = 'y = 2x - 2'
LinePlot(m, b, c, t)

# third line function which is y = mx + b
ax = fig.add_subplot(2, 2, 3)
plotspec(ax)
m = -2
b = 3
c = 'b'
t = 'y = -2x + 3'
LinePlot(m, b, c, t)

# forth line function which is y = mx + b
ax = fig.add_subplot(2, 2, 4)
plotspec(ax)
m = -2
b = -3
c = 'm'
t = 'y = -2x - 3'
LinePlot(m, b, c, t)

# show the plot
show()
```
